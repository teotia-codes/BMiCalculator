```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

#importing dataset using panda
dataset = pd.read_csv('/content/iris_dataset.csv')
#to see what my dataset is comprised of
dataset.head()
```

```python

# Plotting with seaborn
with sns.plotting_context("notebook", font_scale=2.5):
    g = sns.pairplot(dataset[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']],
                     hue='target', palette='tab20', height=6)
g.set(xticklabels=[])
```

```python
from sklearn.model_selection import train_test_split

# Separating independent and dependent variables
X = dataset.iloc[:, :-1].values  # Independent variables
y = dataset.iloc[:, -1].values   # Dependent variable

# Splitting dataset into training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

# Fitting Linear Regression to the training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)

```

```python
print(y_pred)
```

```python

import statsmodels.api as sm
import statsmodels.formula.api as smf
# Rename columns for compatibility with formula notation
dataset.rename(columns={
    'sepal length (cm)': 'sepal_length_cm',
    'sepal width (cm)': 'sepal_width_cm',
    'petal length (cm)': 'petal_length_cm',
    'petal width (cm)': 'petal_width_cm'
}, inplace=True)

# Fitting OLS model
model = smf.ols('target ~ sepal_length_cm + sepal_width_cm + petal_length_cm + petal_width_cm', data=dataset).fit()

# Performing ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Table:")
print(anova_table)
```

```python

# Hypothesis testing
alpha = 0.05  # General significance level
print("Hypothesis Testing:")
for attr in ['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']:
    print(f"Testing {attr}:")
    print(f"Null Hypothesis (H0): The coefficient for {attr} is zero (no effect).")
    print(f"Alternate Hypothesis (H1): The coefficient for {attr} is not zero (effect).")
    print(f"T-value: {model.tvalues[attr]}")
    print(f"P-value: {model.pvalues[attr]}")
    if model.pvalues[attr] < alpha:
        print(f"Result: Reject the null hypothesis for {attr} (p-value < {alpha})")
        print(f"The coefficient for {attr} is statistically significant.\n")
    else:
        print(f"Result: Fail to reject the null hypothesis for {attr} (p-value >= {alpha})")
        print(f"The coefficient for {attr} is not statistically significant.\n")
```

```python
print("T-Test :")
print(model.summary())
```

```python
p_values = model.pvalues
print("P-Values:")
print(p_values)
```

```python

# Separating independent and dependent variables
X = dataset[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']].values
y = dataset['target'].values

# Splitting dataset into training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

# Fitting Linear Regression to the training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)

# Plotting the results
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
plt.scatter(y_test, y_pred, color='red', label='Predicted')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual Target')
plt.ylabel('Predicted Target')
plt.title('Actual vs Predicted Target')
plt.legend()
plt.show()
```

```python
print("New Predicted Values:")
print(y_pred)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

# Number of observations
n = len(y_test)

# Number of predictors
p = X_train.shape[1]

# Calculate Adjusted R2
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"R2: {r2}")
print(f"Adjusted R2: {adjusted_r2}")
```

```python

# Identify the feature with the highest F-score
highest_f_score_feature = anova_table['F'].idxmax()
print(f"Removing feature with highest F-score: {highest_f_score_feature}")

# Remove the feature with the highest F-score
dataset_reduced = dataset.drop(columns=[highest_f_score_feature, 'target'])

# Prepare data for new model
X_new = dataset_reduced.values
y = dataset['target'].values

# Splitting the new dataset into training and testing sets
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y, test_size=0.33, random_state=0)

# Fit the new model to the training data
regressor_new = LinearRegression()
regressor_new.fit(X_train_new, y_train_new)

# Predicting the test set results with the new model
y_pred_new = regressor_new.predict(X_test_new)
```

```python
newr2 = r2_score(y_test_new, y_pred_new)

# Calculate new Adjusted R2
adjusted_r2_new = 1 - (1 - newr2) * (n - 1) / (n - (p - 1) - 1)

print(f"New R2: {newr2}")
print(f"New Adjusted R2: {adjusted_r2_new}")
```

```python
print("Original Model R2:", r2)
print("Original Model Adjusted R2:", adjusted_r2)
print("New Model R2:", newr2)
print("New Model Adjusted R2:", adjusted_r2_new)

if adjusted_r2_new > adjusted_r2:
    print("The new model without the less significant feature is better.")
else:
    print("The original model is better.")
```
